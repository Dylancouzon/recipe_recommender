```python
import kagglehub

# Download recipes dataset from kaggle.com
path = kagglehub.dataset_download("shuyangli94/food-com-recipes-and-user-interactions")

print("Path to dataset files:", path)
```

```python
import pandas as pd  # Importing pandas for data manipulation and analysis
import warnings 
warnings.filterwarnings("ignore")  # Suppressing all warnings to avoid cluttering the output.
```

```python
recipes = pd.read_csv(f"{path}/raw_recipes.csv")
ratings = pd.read_csv(f"{path}/raw_interactions.csv")
```

```python
# Displaying the first 10 rows of the recipes DataFrame
recipes.head(10)
```

```python
ratings.head(10) 
```

```python
# Group by 'recipe_id' and calculate the mean and count of 'rating'
recipe_stats = ratings.groupby('recipe_id')['rating'].agg(['mean', 'count']).reset_index()

# Rename the columns for clarity
recipe_stats.rename(columns={'mean': 'average_rating', 'count': 'rating_count'}, inplace=True)

# Round the average rating to 1 decimal point
recipe_stats['average_rating'] = recipe_stats['average_rating'].round(1)

# Display the result
recipe_stats.head(10)
```

```python
# Merge the recipes dataframe with recipe_stats on recipe_id and id
recipes = recipes.merge(recipe_stats, left_on='id', right_on='recipe_id', how='left')

# Drop the redundant recipe_id column from the merged dataframe
recipes.drop(columns=['recipe_id'], inplace=True)

# Display the updated recipes dataframe
recipes.head(10)
```

```python
import matplotlib.pyplot as plt

# Count the number of recipes for each average_rating
rating_counts = recipes['average_rating'].value_counts().sort_index()

# Plot the bar chart
plt.figure(figsize=(10, 6))
rating_counts.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Number of Recipes by Average Rating', fontsize=16)
plt.xlabel('Average Rating', fontsize=14)
plt.ylabel('Number of Recipes', fontsize=14)

# Change the orientation of x-axis labels
plt.xticks(rotation=90) 

plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Filter recipes with average_rating >= 4 and rating_count >= 5
filtered_recipes = recipes[(recipes['average_rating'] >= 4) & (recipes['rating_count'] >= 5)]

# Display the filtered dataframe
filtered_recipes
```

```python
import openai
import os
from dotenv import load_dotenv

# Load environment variables
load_dotenv()

# Set the OpenAI API key
openai.api_key = os.getenv("OPENAI_API_KEY")

# Define the prompt to generate the list of ingredients
prompt = (
    "Generate a list of the 200 most common used cooking ingredients. focusing strictly on foods, spices, herbs, and pantry staples (no utensils or tools)"
    "Provide the list as a comma-separated string with no additional message."
)

# initially used gtp-4.1-nano for speed and cost efficiency, but switched to gpt-4.1 for better accuracy.
# 100 ingredients = Number of recipes with unlisted ingredients: 46088
response = openai.responses.create(
    model="gpt-4.1",
    instructions="You are a helpful cooking assistant.",
    input=prompt,
    temperature=0.3,

)

response.status # Check if the response is successful
```

```python
import csv

# Extract the text from the response
ingredients_text = response.output[0].content[0].text

# Split the text into a list of ingredients
ingredients_list = [ingredient.strip() for ingredient in ingredients_text.split(',')]

# Define the output CSV file path
csv_file_path = 'output_data/ingredients.csv'

# Write the ingredients to the CSV file
with open(csv_file_path, mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['Ingredient'])  # Add a header row
    for ingredient in ingredients_list:
        writer.writerow([ingredient])

print(f"Ingredients have been saved to {csv_file_path}")
```

```python
num_recipes = len(filtered_recipes)
print(f"Number of recipes: {num_recipes}")
```

```python
# Load the ingredients from the CSV file
ingredients_from_csv = pd.read_csv(csv_file_path)['Ingredient'].str.lower().tolist()

#Create a function to check if a recipe has unlisted ingredients
def has_unlisted_ingredient(recipe):
   # Turn the string into an array
   l_clean = recipe['ingredients'].strip("[]")
   ingredients_list = [item.strip().strip("'\"") for item in l_clean.split(',')]
   return set(ingredients_list).issubset(ingredients_from_csv)

unlisted_recipes = filtered_recipes[~filtered_recipes.apply(has_unlisted_ingredient, axis=1)]
print("Number of recipes:", len(filtered_recipes))
print("Number of recipes with unlisted ingredients:", len(unlisted_recipes))
```

```python
from collections import Counter

# Flatten the list of ingredients in unlisted_recipes
unlisted_ingredients = [ingredient.strip().lower() for ingredients in unlisted_recipes['ingredients'] for ingredient in eval(ingredients)]

# Filter out ingredients that are already in ingredients_from_csv
filtered_ingredients = [ingredient for ingredient in unlisted_ingredients if ingredient not in ingredients_from_csv]

# Count the occurrences of each ingredient
ingredient_counts = Counter(filtered_ingredients)

# Get the top 20 ingredients
top_20_unlisted_ingredients = ingredient_counts.most_common(20)

print(top_20_unlisted_ingredients)

```

```python
# We can see that a lot of the top 20 unlisted ingredients are actually different names for the same ingredients listed in the CSV file.
# This function appends the top 20 items to ingredients.csv to increase the number of recipes we can use.

with open(csv_file_path, mode='a', newline='') as file:
    writer = csv.writer(file)
    for ingredient, count in top_20_unlisted_ingredients:
        writer.writerow([ingredient])
print("Top 20 unlisted ingredients have been added to the CSV file.")
```

```python
# Reload the ingredients from the CSV file to include the new ones
ingredients_from_csv = pd.read_csv(csv_file_path)['Ingredient'].str.lower().tolist()

common_ingredients_recipes = filtered_recipes[filtered_recipes.apply(has_unlisted_ingredient, axis=1)]
print("Total number of recipes:", len(filtered_recipes))

# The number of recipes with common ingredients jumped from 330 to 1025 with the most recent change
print("Number of recipes with common ingredients:", len(common_ingredients_recipes))
```

```python
## Saving the common ingredients recipes to a CSV file. 

# Define the output CSV file path
common_recipes_csv_path = 'output_data/common_ingredients_recipes.csv'

# Drop unnecessary columns
fields_to_drop = ['contributor_id', 'submitted', 'nutrition', 'rating_count', 'has_unlisted_ingredient']
common_ingredients_recipes = common_ingredients_recipes.drop(columns=fields_to_drop)

# Export the DataFrame to a CSV file

common_ingredients_recipes.to_csv(common_recipes_csv_path, index=False)

print(f"Common ingredients recipes have been saved to {common_recipes_csv_path}")
```
